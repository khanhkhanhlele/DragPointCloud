```python
import numpy as np
import torch
path = "point_cloud_no_drag.pt"
pc = torch.load(path)
print(pc.shape)
pc = pc.cpu().numpy().reshape((2048, 3))
#plot point cloud using plotly.graph_objs
import plotly.graph_objs as go
# pc = res[0]

def plot_3D(pc, axis_labels=False):
    x, y, z = pc[:, 0], pc[:, 1], pc[:, 2]
    fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers', marker=dict(size=2))])
    if axis_labels:
        fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
    fig.show()

plot_3D(pc, axis_labels=True)
```

```python
import plotly.graph_objs as go

def plot_3D(pc, axis_labels=False):
    fig_plotly = go.Figure(
        data=[
            go.Scatter3d(
                x=pc[:, 0], y=pc[:, 1], z=pc[:, 2],
                mode='markers',
                marker=dict(
                    size=2,
                ),
                # Thm hovertext  hin th index ca mi im
                hovertext=[f"Index: {i}<br>Pos: ({x:.2f}, {y:.2f}, {z:.2f})" 
                            for i, (x, y, z) in enumerate(pc)],
                hoverinfo='text',  # Hin th hover text khi hover
            )
        ],
        layout=dict(
            scene=dict(
                xaxis=dict(visible= axis_labels),
                yaxis=dict(visible= axis_labels),
                zaxis=dict(visible= axis_labels)
            )
        ),
    )
    # Hin th  th
    fig_plotly.show(renderer="vscode")
```

```python
plot_3D(pc)
```
